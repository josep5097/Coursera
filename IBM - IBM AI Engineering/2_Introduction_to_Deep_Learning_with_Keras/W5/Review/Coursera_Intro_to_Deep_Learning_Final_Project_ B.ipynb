```python
import pandas as pd
import numpy as np
```

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
#Splitting the data
predictors = concrete_data[['Cement', 'Blast Furnace Slag', 'Fly Ash', 'Water', 'Superplasticizer', 'Coarse Aggregate', 'Fine Aggregate', 'Age', 'Strength']]
target = concrete_data['Strength']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3)
```

```python
n_cols = predictors.shape[1]
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

def regression_model():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))

    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
model = regression_model()
```

```python
#PART B
#Mean of MSE (part A):  27.483412318318827
#Standard Deviation of MSE part A): 165.79596712022925
```

```python
#Normalizing data
predictors_norm = (predictors - predictors.mean()) / predictors.std()
X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.3)
X_train
```

```python
model.fit(X_train, y_train, validation_split=0.3, epochs=50, verbose=2)
```

```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X_test)
mean_squared_error(y_test, y_pred)
```

```python
#Iterating 50 times
mse = []
model = regression_model()

for i in range(50):
    print('Iteration number: ' + str(i))
    X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3)
    model.fit(X_train, y_train, validation_split=0.3, epochs=50, verbose=2)
    y_pred = model.predict(X_test)
    mse.append(mean_squared_error(y_test, y_pred))
```

```python
mse
```

```python
mean_B = np.mean(mse)
print('Mean of MSE: ', mean_B)

std_B = np.std(mse)
print('Standard Deviation of MSE:', std_B)
```
