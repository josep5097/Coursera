```
import keras
import pandas as pd

import numpy as np

from keras.models import Sequential
from keras.layers import Dense

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```
concrete_data = pd.read_csv('https://cocl.us/concrete_data')
concrete_data.head()
```

```
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column

n_cols = predictors.shape[1]
```

```
# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

# Part A : baseline model

```
errors = np.zeros((50,))
for ik in range(50):

  X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3)
  model = regression_model()
  model.fit(X_train, y_train, epochs=50, verbose=2)
  y_predict = model.predict(X_test)

  errors[ik] = mean_squared_error(y_test, y_predict)

print(errors)
```

```
print("Mean of squared errors : %f" % (np.mean(errors)))
print("Stderr of squared errors : %f" % (np.std(errors,ddof=1)))
```

# Part B : normalized model

```
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```
errors_norm = np.zeros((50,))
for ik in range(50):

  X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.3)
  model = regression_model()
  model.fit(X_train, y_train, epochs=50, verbose=2)
  y_predict = model.predict(X_test)

  errors_norm[ik] = mean_squared_error(y_test, y_predict)

print(errors_norm)
```

```
print("Mean of squared errors : %f" % (np.mean(errors_norm)))
print("Stderr of squared errors : %f" % (np.std(errors_norm,ddof=1)))
```

Normalizing the errors allows to decrease the standard error, the model is allowed to learn homogeneously among all the data.
The mean of squared errors does not change dramatically

# Part C : number of epochs

```
errors_epochs = np.zeros((50,))
for ik in range(50):

  X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.3)
  model = regression_model()
  model.fit(X_train, y_train, epochs=100,verbose=2)
  y_predict = model.predict(X_test)

  errors_epochs[ik] = mean_squared_error(y_test, y_predict)

print(errors_epochs)
```

```
print("Mean of squared errors : %f" % (np.mean(errors_epochs)))
print("Stderr of squared errors : %f" % (np.std(errors_epochs,ddof=1)))
```

Having increased the number of epochs leads to a strong decrease in mean and standard errors, showing the system was not fully converged with 50 epochs yet !

# Part D : hidden layers

```
# define regression model
def regression_model_more_layers():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```
errors_hidden = np.zeros((50,))
for ik in range(50):

  X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.3)
  model = regression_model_more_layers()
  model.fit(X_train, y_train, epochs=50, verbose=2)
  y_predict = model.predict(X_test)

  errors_hidden[ik] = mean_squared_error(y_test, y_predict)

print(errors_hidden)
```

```
print("Mean of squared errors : %f" % (np.mean(errors_hidden)))
print("Stderr of squared errors : %f" % (np.std(errors_hidden,ddof=1)))
```

The error decrease in average but the standard error increase a lot, being a consequence of overfitting, the model being too complex for the small number of training data
